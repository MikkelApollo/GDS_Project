```python
import pandas as pd
```

```python
fnc = pd.read_csv('FNC_lemmatized.csv')

fnc.content.str.strip("]['").str.split("', '")
```

```python
types = fnc.type.tolist()
  
type_dict = {key: types.count(key) for key in types}

print(type_dict)
```

```python
fake_num = type_dict.get('unreliable') + type_dict.get('fake') + type_dict.get('conspiracy') + type_dict.get('bias') + type_dict.get('junksci') + type_dict.get('satire')
reli_num = type_dict.get('clickbait') + type_dict.get('political') + type_dict.get('reliable')

print('Fake articles       : {}'.format(fake_num))
print('Reliable articles   : {}'.format(reli_num))
print('Fake percentage     : {}'.format(fake_num/(fake_num+reli_num)*100))
print('Reliable percentage : {}'.format(reli_num/(fake_num+reli_num)*100))
```

```python
domains = fnc.domain.tolist()

domain_dict = {key: domains.count(key) for key in domains}

sorted_domain_dict = sorted(domain_dict.items(), key=lambda x:x[1], reverse=True)
```

```python
def get_nums(lst):
  nums = 0
  for word in lst:
    if word == 'num':
      nums += 1
  return nums

def get_urls(lst):
  urls = 0
  for word in lst:
    if word == 'url':
      urls += 1
  return urls

def get_dates(lst):
  dates = 0
  for word in lst:
    if word == 'date':
      dates += 1
  return dates

fnc['nums'] = fnc['content'].apply(get_nums)
fnc['urls'] = fnc['content'].apply(get_urls)
fnc['dates'] = fnc['content'].apply(get_dates)
```

```python
type_df = pd.DataFrame()

type_df['types'] = ('political', 'fake', 'satire', 'reliable', 'conspiracy', 'unreliable', 'bias', 'rumor', 'clickbait', 'junksci')

type_df['nums'] = fnc.loc[fnc['type'] == type_df['type'], 'nums'].sum()
type_df['urls'] = fnc.loc[fnc['type'] == type_df['type'], 'urls'].sum()
type_df['dates'] = fnc.loc[fnc['type'] == type_df['type'], 'dates'].sum()

type_df.loc['political', 'nums':'dates'].div(194518)
type_df.loc['fake', 'nums':'dates'].div(104883)
type_df.loc['satire', 'nums':'dates'].div(13160)
type_df.loc['reliable', 'nums':'dates'].div(218564)
type_df.loc['conspiracy', 'nums':'dates'].div(97314)
type_df.loc['unreliable', 'nums':'dates'].div(35332)
type_df.loc['bias', 'nums':'dates'].div(133232)
type_df.loc['rumor', 'nums':'dates'].div(56445)
type_df.loc['clickbait', 'nums':'dates'].div(27412)
type_df.loc['junksci', 'nums':'dates'].div(14040)
```
